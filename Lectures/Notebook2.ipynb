```python
datadir = "data/"
import numpy as np
import pandas as pd
pd.set_option('display.max.columns', 100)
# to draw pictures in jupyter notebook
%matplotlib inline 
import matplotlib.pyplot as plt
import seaborn as sns
# we don't like warnings
# you can comment the following 2 lines if you'd like to
import warnings
warnings.filterwarnings('ignore')
```

```python
telecom_churn = pd.read_csv(datadir + 'telecom_churn.csv')
```

```python
telecom_churn
```

```python
telecom_churn.shape
```

```python
telecom_churn["International plan"] = telecom_churn["International plan"].map({"Yes" : 1, "No" : 0})
telecom_churn["Voice mail plan"] = telecom_churn["Voice mail plan"].map({"Yes" : 1, "No" : 0})
```

```python
telecom_churn["Churn"] = telecom_churn["Churn"].astype("int")
```

```python
plt.rcParams["figure.figsize"] = (20, 20)
telecom_churn.drop(["State"], axis = 1).hist();
```

```python
telecom_churn.corr()
```

```python
plt.rcParams["figure.figsize"] = (16, 12)
sns.heatmap(telecom_churn.corr());
```

```python
# drop all charge fields
telecom_churn.drop([feature for feature in telecom_churn.columns if "charge" in feature], axis = 1, inplace = True)
telecom_churn.shape
```

## Interactions between features

### numeric-numeric

```python
# save & drop state
states = telecom_churn.State
telecom_churn.drop("State", axis = 1, inplace = True)
```

```python
# use corr to study num - num
telecom_churn.corrwith(telecom_churn["Total day minutes"])
```

```python
# do two nums in plot
plt.scatter(telecom_churn["Total day minutes"], telecom_churn["Customer service calls"]);
```

### categorical-categorical

```python
# even though telecom_churn["Customer service calls"], is actually ordinal, thus categ
pd.crosstab(telecom_churn["Churn"], telecom_churn["Customer service calls"])
```

```python
sns.countplot(x = "Customer service calls", hue = "Churn", data = telecom_churn)
plt.title("Customer service calls stay vs churn");
```

### categorical - numeric

```python
telecom_churn.groupby("Churn")["Total day minutes", "Customer service calls"].mean()
```

```python
telecom_churn.groupby("Churn")["Total day minutes", "Customer service calls"].agg([np.mean, np.std])
```

```python
plt.rcParams["figure.figsize"] = (8, 6)
sns.boxplot(x = "Churn", y = "Total day minutes", data = telecom_churn);
```

## TSNE

```python
from sklearn.manifold import TSNE
```

```python
telecom_churn_tsne = TSNE(random_state=17)
```

```python
#telecom_churn
```

```python
%%time
X_repr_all = telecom_churn_tsne.fit_transform(telecom_churn)
```

```python
#X_repr_churn = telecom_churn_tsne.fit_transform(telecom_churn[telecom_churn["Churn"] == 1])
```

```python
X_repr_all.shape
```

```python
chartdir = "charts/"
plt.rcParams["figure.figsize"] = (16, 12)
plt.scatter(X_repr_all[telecom_churn["Churn"] != 1, 0],
            X_repr_all[telecom_churn["Churn"] != 1, 1],
            label = "stay")
plt.scatter(X_repr_all[telecom_churn["Churn"] == 1, 0],
            X_repr_all[telecom_churn["Churn"] == 1, 1],
            c = 'r', label = "churn")
plt.legend()
plt.xlabel("tSNE axis 1")
plt.ylabel("tSNE axis 2")
plt.title("tSNE Representation")
plt.savefig(chartdir + "tsne_churn.png", dpi = 150)
```
