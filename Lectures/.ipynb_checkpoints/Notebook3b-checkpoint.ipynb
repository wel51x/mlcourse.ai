```python
%cd /notebooks
datadir = "data/"
import numpy as np
import pandas as pd
pd.set_option('display.max.columns', 100)
# to draw pictures in jupyter notebook
%matplotlib inline 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV
from sklearn.tree import export_graphviz
# we don't like warnings
# you can comment the following 2 lines if you'd like to
import warnings
warnings.filterwarnings('ignore')
```

```python
telecom_churn = pd.read_csv(datadir + 'telecom_churn.csv')
```

```python
telecom_churn
```

```python
telecom_churn.info()
```

### preprocess

```python
telecom_churn["International plan"] = telecom_churn["International plan"].map({"No" : 0, "Yes" : 1})
telecom_churn["Voice mail plan"] = telecom_churn["Voice mail plan"].map({"No" : 0, "Yes" : 1})
telecom_churn["Churn"] = telecom_churn["Churn"].astype("int")
states = telecom_churn.pop("State")  # remove and store in series
```

```python
telecom_churn
```

```python
telecom_churn.describe()
```

### Split telecom_churn into vector y (churn) and matrix X (all else)

```python
X, y = telecom_churn.drop("Churn", axis = 1), telecom_churn.Churn
```

```python
X.shape, y.shape
```

```python
X_train, X_holdout, y_train, y_holdout = train_test_split(X, y, test_size=.25, random_state=17)
```

```python
X_train.shape, X_holdout.shape
```

### Decision Tree fit - predict

```python
# instantiate dec tree with NO params
tree = DecisionTreeClassifier(random_state=17)
```

```python
# fit
tree.fit(X_train, y_train)
```

```python
# predict
pred_holdout = tree.predict(X_holdout)
pred_holdout.shape, y_holdout.shape
```

```python
# check accuracy
original_score = accuracy_score(y_holdout, pred_holdout)
print("Accuracy =", original_score)
```

```python
# what pct churn?
y.value_counts(normalize = True)
```

### Hyperparameter tuning

```python
param_grid = {"max_depth" : np.arange(2, 11), "min_samples_leaf" : np.arange(1, 11)}
```

```python
# use StratifiedKFold object for cross-validation param to GridSearchCV
skf = StratifiedKFold(n_splits = 5, shuffle = True, random_state=17)
```

```python
# get best tree
best_tree = GridSearchCV(tree, param_grid, cv=skf, n_jobs = -1, verbose = 1)
```

```python
# and fit it
best_tree.fit(X_train, y_train)
```

```python
# what are the params?
print("best_params_ =", best_tree.best_params_, "\nbest_estimator_ =", best_tree.best_estimator_)
```

### Cross-validation assessment of model accuracy

```python
# and what's its score?
print("best_score_ =", best_tree.best_score_)
```

### Now test against holdout data

```python
# predict
pred_holdout = best_tree.predict(X_holdout)
# and check accuracy
best_score = accuracy_score(y_holdout, pred_holdout)
print("Accuracy =", best_score)
```

```python
print("Improvement over original =", (best_score - original_score) * 100, "percent")
```

### For tree visualization

```python
# instantiate dec tree with max_depth=3
tree = DecisionTreeClassifier(max_depth=3, random_state=17)
# fit
tree.fit(X_train, y_train)
```

```python
from io import StringIO
import pydotplus
from ipywidgets import Image
```

```python
dot_data = StringIO()
export_graphviz(decision_tree=tree,
                out_file=dot_data,
                filled=True,
                feature_names=telecom_churn.drop("Churn", axis = 1).columns)
```

```python
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
Image(value=graph.create_png())
```
